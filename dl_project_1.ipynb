**Extracting the images from the zip file**

```python
from zipfile import ZipFile
with ZipFile("dog_cat_images.zip", "r") as zip:
    zip.extractall()
```

**Importing dependencies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import cv2
import os
```

**Resizing the images**

```python
# Directory for the dog and cat images
dog_cat_img = "dog_cat_images"
```

```python
# New directory for resized images
resized_img = "resized_images"
os.makedirs(resized_img, exist_ok=True)
```

```python
# Resizing the images
for i in os.listdir(dog_cat_img):
    # print(i)
    # Create a path for the images
    img_path = os.path.join(dog_cat_img, i)
    # print(img_path)
    # Resize the images
    images = Image.open(img_path)
    resize = images.resize((224, 224))
    resize = resize.convert("RGB")
    # Create a new path for the resized images
    resize_img_path = os.path.join(resized_img, i)
    # Saving the resized images in the resized images folder
    resize.save(resize_img_path)
```

**Getting the shape of a resized image**

```python
img = cv2.imread("resized_images\cat.12.jpg")
img = plt.imshow(img)
plt.show()
```

```python
img.get_shape()
```

```python
# Converting the images to labels
labels = []
for img in os.listdir(resized_img):
    # print(img)
    img = img[0:3]
    # print(img)
    if img == "cat":
        labels.append(0)
    else:
        labels.append(1)
print(labels)
```

```python
# Counting the number of dog and cat images
values, counts = np.unique(labels, return_counts=True)
print(values)
print(counts)
```

**Converting the labels to numpy arrays**

```python
import cv2
import glob
image_extensions = ["png", "jpg"]
files = []

for e in image_extensions:
    for file in glob.glob(os.path.join(resized_img, f"*.{e}")):
        files.append(file)
print(files)
```

```python
# Converting the images to a numpy array
resized_img_array = np.asarray([cv2.imread(file) for file in files])
resized_img_array
```

**Train Test Split**

```python
# Separating the data into features and target
X = resized_img_array
Y = np.asarray(labels)
```

```python
# Splitting the data into training and test data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape)
```

**Scaling the data**

```python
X_train_scaled = X_train/255
X_test_scaled = X_test/255
```

```python
X_train_scaled
```

**Building the Neural Network**

```python
# Importing the DL libraries
import tensorflow as tf # Used for building and training models
import tensorflow_hub as hub # Used for loading pre-trained model
```

```python
# Getting the pretrained model(MobileNetV2)
mobilenet_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
```

```python
# Loading the pretrained model
pretrained_model = hub.KerasLayer(mobilenet_model, input_shape=(224,224,3), trainable=False)
```

```python
# Creating the layers of the neural network
no_of_classes = 2
model = tf.keras.Sequential([
    pretrained_model,
    tf.keras.layers.Dense(no_of_classes)
])
model.summary()
```

**Compiling the model**

```python
model.compile(
    optimizer="adam",
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=["acc"]
)
```

```python
model.fit(X_train_scaled, Y_train, epochs=5)
```

**Model Evaluation**

```python
score, acc = model.evaluate(X_test_scaled, Y_test)
print(f"Loss: {score}\nAccuracy: {acc}")
```

**Predictive system**

```python
# Accepting user input
input_image_path = input("Path of the image to be predicted: ")
# Loading the image 
input_image = cv2.imread(input_image_path)
# Resizing the image
input_image_resize = cv2.resize(input_image, (224, 224))
# Scaling the data
input_scaled = input_image_resize/255
# Reshaping the image as an instance
image_reshape = np.reshape(input_scaled, [1, 224, 224, 3])
# Making prediction
input_prediction = model.predict(image_reshape)
# Getting the maximum value for prediction
input_label = np.argmax(input_prediction)
# Condition for prediction  
if input_label == 0:
    print("This is a cat image")
else:
    print("This is a dog image")
```
